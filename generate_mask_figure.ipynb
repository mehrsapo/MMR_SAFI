```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
from matplotlib.colors import LogNorm
from pylab import figure, cm
```

```python
mm_mask = np.load("fig_data/mask_average_mdrr_25.npy")
mm_Ly = np.load("fig_data/Ly_mdrr_25.npy")
mm_Ly_masked = np.load("fig_data/Ly_masked_mdrr_25.npy")

sf_mask = np.load("fig_data/mask_average_safi_25.npy")
sf_Ly = np.load("fig_data/Ly_safi_25.npy")
sf_Ly_masked = np.load("fig_data/Ly_safi_mdrr_25.npy")
```

```python
plt.figure(figsize=(20, 20))
vmin_= 0; vmax_=1 
plt.subplot(161)
plt.imshow(mm_mask / mm_mask.max(), cmap='gray', vmin=vmin_, vmax = vmax_); plt.axis('off')
plt.title('$\sum_{c=1}^{N_C} \mathbf{\Lambda}_c({\mathbf{x}}_{\mathrm{MMR}})$')
plt.subplot(162)
plt.imshow(mm_Ly / mm_Ly.max(), cmap='gray', vmin=vmin_, vmax = vmax_); plt.axis('off')
plt.title('$\sum_{c=1}^{N_C} \\vert \mathbf{W}_c \mathbf{y} \\vert$')
plt.subplot(163)
plt.imshow(mm_Ly_masked / mm_Ly_masked.max(), cmap='gray', vmin=vmin_, vmax = vmax_); plt.axis('off')
plt.title('$\sum_{c=1}^{N_C} (\mathbf{\Lambda}_c({{\mathbf{x}}}_{\mathrm{MMR}}) \odot \\vert \mathbf{W}_c \mathbf{y} \\vert)$')
plt.subplot(164)
plt.imshow(sf_mask / sf_mask.max(), cmap='gray', vmin=vmin_, vmax = vmax_); plt.axis('off')
plt.title('$\sum_{c=1}^{N_C} \\tilde{\mathbf{\Lambda}}_c({\mathbf{x}}_{\mathrm{SAFI}})$')
plt.subplot(165)
plt.title('$\sum_{c=1}^{N_C} \\vert \mathbf{W}_c \mathbf{y} \\vert$')
plt.imshow(sf_Ly / sf_Ly.max(), cmap='gray', vmin=vmin_, vmax = vmax_); plt.axis('off')
plt.subplot(166)
plt.imshow(sf_Ly_masked / sf_Ly_masked.max(), cmap='gray', vmin=vmin_, vmax = vmax_); plt.axis('off')
plt.title('$\sum_{c=1}^{N_C}(\\tilde{\mathbf{\Lambda}}_c({\mathbf{x}}_{\mathrm{SAFI}}) \odot \\vert \mathbf{W}_c \mathbf{y} \\vert)$')
plt.subplots_adjust(wspace=0.05, hspace=0.2)
plt.savefig('masks_both.pdf', bbox_inches="tight")
plt.show()
```
